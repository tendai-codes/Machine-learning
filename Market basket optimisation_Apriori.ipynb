```
from apyori import apriori
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

## Data Preprocessing

```
dataset = pd.read_csv('Market_Basket_Optimisation.csv', header = None)
transactions = []
for i in range(0, 7501):
  transactions.append([str(dataset.values[i,j]) for j in range(0, 20)])
```

- Line 3 > we use a for loop to creat and populate the 'transcations' list
- To do this we use a for loop gooing through all the rows of the data (0-7501).
- Line 4 comands the for loop to populate list with items from each row using the (i) function, while we use (j) to populate the list for all of the rows  
- str - string values because data is items not numbers

## Training the Apriori model on the dataset

```
rules = apriori(transactions = transactions, min_support = 0.003, min_confidence = 0.2, min_lift = 3, min_length = 2, max_length = 2)
```

- transactions parameter is the 'transaction' list we created prior
- min_support > (number of times to see item * time period)/total items [transaction] [(3*7)/7501]
- min_confidence = portion of time rule is correct (there is no rule of thumb here
- min_lift generally anything above 3
- min_length = 2
- max_length depends on outcoume eg buy 1 product, get 1 free > (1+1) = 2, and buy 2 products,, get 1 free > (2+1) = 3

### Putting the results well organised into a Pandas DataFrame

```
def inspect(results):
    lhs         = [tuple(result[2][0][0])[0] for result in results]
    rhs         = [tuple(result[2][0][1])[0] for result in results]
    supports    = [result[1] for result in results]
    confidences = [result[2][0][2] for result in results]
    lifts       = [result[2][0][3] for result in results]
    return list(zip(lhs, rhs, supports, confidences, lifts))
resultsinDataFrame = pd.DataFrame(inspect(results), columns = ['Left Hand Side', 'Right Hand Side', 'Support', 'Confidence', 'Lift'])
```

- lhs > left hand side
- rhs > right hand side
